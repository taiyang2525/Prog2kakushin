# 〇〇県（府，都）の気象情報取得

コードを一番上のセルから実行してください．一番上のセルを実行すると，都道府県の一覧がズラリと表示されます．

```python
import pandas as pd
import requests
import sys


#ファイルを総務省統計局のサイトからダウンロード
url = "https://www.stat.go.jp/data/nihon/zuhyou/n230200200.xlsx"
response = requests.get(url)
#200なら成功
if response.status_code == 200:
    with open('prefecture.xlsx', 'wb') as file:
      file.write(response.content)
    print("ファイルのダウンロードが完了しました")
else:
    print("ファイルのダウンロードに失敗しました")
    sys.exit()

#dfの処理
df = pd.read_excel('prefecture.xlsx')
# NaNを含む行を取り除く（axis=0は行、axis=1は列を指定）
df = df.dropna(axis=0)  # NaNを含む行を削除
df = df[df['2-2\u3000都道府県別人口と人口増減率'] != "全国"]
df = df[['2-2\u3000都道府県別人口と人口増減率']]
df = df.rename(columns={'2-2\u3000都道府県別人口と人口増減率': '都道府県'})
df = df.reset_index()
df = df.drop('index', axis=1)
display(df)
```

気になった都道府県の番号（0～46）を入力してください．気象情報を取得します．

```python
import requests
import xml.etree.ElementTree as ET
import os

#1-47の整数値を入力させる処理

while True:
  k = input('0～46の整数値で対応する都道府県のものを入力してください：')
  if k.isdigit() == False:
    print('再度入力してください')
    continue
  else:
    k = int(k)
    if 0 <= k <= 46:
      break
    else:
      print('再度入力してください')
      continue
#indexから都道府県名に変換

pref_cap =  {
    '北海道': '札幌',
    '青森': '青森',
    '岩手': '盛岡',
    '宮城': '仙台',
    '秋田': '秋田',
    '山形': '山形',
    '福島': '福島',
    '茨城': '水戸',
    '栃木': '宇都宮',
    '群馬': '前橋',
    '埼玉': 'さいたま',
    '千葉': '千葉',
    '東京': '東京',
    '神奈川': '横浜',
    '新潟': '新潟',
    '富山': '富山',
    '石川': '金沢',
    '福井': '福井',
    '山梨': '甲府',
    '長野': '長野',
    '岐阜': '岐阜',
    '静岡': '静岡',
    '愛知': '名古屋',
    '三重': '津',
    '滋賀': '大津',
    '京都': '京都',
    '大阪': '大阪',
    '兵庫': '神戸',
    '奈良': '奈良',
    '和歌山': '和歌山',
    '鳥取': '鳥取',
    '島根': '松江',
    '岡山': '岡山',
    '広島': '広島',
    '山口': '山口',
    '徳島': '徳島',
    '香川': '高松',
    '愛媛': '松山',
    '高知': '高知',
    '福岡': '福岡',
    '佐賀': '佐賀',
    '長崎': '長崎',
    '熊本': '熊本',
    '大分': '大分',
    '宮崎': '宮崎',
    '鹿児島': '鹿児島',
    '沖縄': '那覇'
}
pref = pref_cap[df['都道府県'][k]]


city_name = pref

xml_url = 'https://weather.tsukumijima.net/primary_area.xml'
base_url ='https://weather.tsukumijima.net/api/forecast/'

xml_file = requests.get(xml_url).text
root = ET.fromstring(xml_file)
city_id_dict={}
for value in root.iter('city'):
    city_id_dict[value.attrib['title']] = value.attrib['id']

json_file = requests.get(os.path.join(base_url,'city',city_id_dict[city_name])).json()
print(city_name + 'の今日の天気は, ' + json_file['forecasts'][0]['telop'] + 'です．')
print(json_file['description']['text'])
```
