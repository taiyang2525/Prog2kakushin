## 行列１の読み込み

```python
from google.colab import files
import pandas as pd
import numpy as np


def calc(frame):
  global matr
  array = frame.to_numpy()
  matr = np.matrix(array)
  if matr.shape[0] == 0 and matr.shape[1] == 0:
    print('ExcelファイルまたはCSVファイルに情報を入力してください．')
    uploader()
  elif matr.shape[0] == matr.shape[1]:
    display(frame)
    print(str(matr.shape[0]) + "次正方行列を読み込みました．")
    print('（空白セルは0で置換されます．）')
  else:
    display(frame)
    print(str(matr.shape[0]) + "行" + str(matr.shape[1]) + "列の行列を読み込みました．")
    print('（空白セルは0で置換されます．）')




def csv_reader(file):
  df = pd.read_csv(file, header = None)
  df = df.fillna(0) #NaNを0に置き換え
  is_numeric = df.applymap(lambda x: isinstance(x, (int, float))).all().all()
  if is_numeric:
    calc(df)
  else:
    print('ファイルに文字列が含まれます．数値のみであるかを確認してください．')
    uploader()

def excel_reader(file):
  df = pd.read_excel(file, header = None)
  df = df.fillna(0) #NaNを0に置き換え
  is_numeric = df.applymap(lambda x: isinstance(x, (int, float))).all().all()
  if is_numeric:
    calc(df)
  else:
    print('ファイルに文字列が含まれます．数値のみであるかを確認してください．')
    uploader()

def uploader():
  print('行列データをExcelファイル，またはCSVファイルで読み込んでください')
  while True:
    uploaded = files.upload()
    f_name = list(uploaded.keys())[0]
    if '.xlsx' in f_name:
      print('Excelファイルを読み込みました．')
      excel_reader(f_name)
      break
    elif '.csv' in f_name:
      print('CSVファイルを読み込みました．')
      csv_reader(f_name)
      break
    else:
      print('対応していないファイルフォーマットです．やり直してください．')

uploader()
```

## 行列1と行列2の演算

```python
def mode_select():
  global mode
  print('行列1と行列2で演算を行います．')
  while True:
    print('--------------------------------------')
    print('加法：1')
    print('減法：2')
    print('乗法：3')
    print('--------------------------------------')
    mode= input('数字を入力してください：')
    if mode == "1" or mode == "2" or mode == "3":
      mode = int(mode)
      break
    else:
      print('再度入力してください')

print('行列１：入力された行列')
display(pd.DataFrame(matr))
print('行列２：入力された行列')
display(pd.DataFrame(matr2))
mode_select()

if mode == 1 or mode == 2:
  if matr.shape[0] == matr2.shape[0] and matr.shape[1] == matr2.shape[1]:
    if mode == 1:
      matr_new = matr + matr2
      print('計算結果：')
      display(pd.DataFrame(matr_new))
    if mode == 2:
      matr_new = matr - matr2
      print('計算結果：')
      display(pd.DataFrame(matr_new))
  else:
    print('2つの行列の行数・列数が等しいときのみ、加法・減法は定義されます。')

elif mode == 3:
  if matr.shape[1] == matr2.shape[0]:
      matr_new = matr @ matr2
      print('計算結果：')
      display(pd.DataFrame(matr_new))
  else:
    print('行列１の列数と行列２の行数が等しいときのみ，積が定義されます．')
```

## 行列1の計算

行列のスカラー倍・n乗

```python
def mode_select():
  global mode
  print('行列1の演算を行います．')
  while True:
    print('--------------------------------------')
    print('スカラー倍：1（行列１にスカラーcを掛ける）')
    print('n乗：2（行列1をn乗します）')
    print('--------------------------------------')
    mode = input('数字を入力してください：')
    if mode == "1" or mode == "2":
      mode = int(mode)
      break
    else:
      print('再度入力してください')


print('行列１：入力された行列')
display(pd.DataFrame(matr))
mode_select()
if mode == 1:
  c = input('スカラーcの値を入力してください，')
  if c.isnumeric() == True:
    matr_new = float(c) * matr
    print('計算結果：')
    display(pd.DataFrame(matr_new))
  else:
    print('cには数値を入力してください．')
else:
  n = input('指数nの値を入力してください，')
  if n.isnumeric() == True:
    if float(n) == int(n):
      matr_new = np.matrix(matr, dtype = object)
      matr_new = matr_new ** int(n)
      print('計算結果：')
      display(pd.DataFrame(matr_new))
    else:
      print('整数値を入力してください．')
  else:
    print('指数nには数値を入力してください．')
```

行列１の逆行列を求める．

```python
print('入力された行列')
display(pd.DataFrame(matr))
if matr.shape[0] != matr.shape[1]:
  print('正方行列を読み込んでください')
elif np.linalg.det(matr) == 0:
  print('行列式＝０より，正則でないため逆行列を持ちません．')
else:
  print('逆行列はこちらです')
  display(pd.DataFrame(np.linalg.inv(matr)))
```

行列１の行列式を求める

```python
print('入力された行列')
display(pd.DataFrame(matr))
if matr.shape[0] != matr.shape[1]:
  print('正方行列を読み込んでください')
else:
  print('行列式の値（det）： ', np.linalg.det(matr))
```

行列１の固有値を求める（同一の数が出る場合は重解を意味します．）

```python
print('入力された行列')
display(pd.DataFrame(matr))
if matr.shape[0] != matr.shape[1]:
  print('正方行列を読み込んでください')
else:
  λ = np.linalg.eig(matr)[0]
  print('固有値は以下の通りです．')
  print(λ)
```

階数を求める

```python
print('入力された行列')
display(pd.DataFrame(matr))
print('階数 (rank)：', np.linalg.matrix_rank(matr))
```

トレース（対角成分の和）を求める

```python
print('入力された行列')
display(pd.DataFrame(matr))
print('計算結果：', np.trace(matr))
```
